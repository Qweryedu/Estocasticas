```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def histogram(s, nbins=10, width=.5, title='Histograma', color='teal'):
    intervals = np.linspace( s.min(), s.max(), nbins+1 )
    
    left = s[:,np.newaxis] > intervals[:-1][np.newaxis]
    right = s[:,np.newaxis] < intervals[1:][np.newaxis]
    
    hist = (left*right).sum(axis=0)
    
    plt.figure(figsize=(15,5))
    plt.bar(intervals[:-1], hist, width, color=color)
    plt.grid()
    plt.title(title, size=25, color='mediumseagreen')
    plt.xticks(color='slategrey')
    plt.yticks(color='slategrey')
    plt.show()
    
    return intervals[:-1], hist
```

```python
def graph2(intx, histx, inty, histy, colorx='deeppink', colory='blueviolet'):
    plt.figure(figsize=(15,5))
    #Se grafican los datos de x
    plt.bar(intx, histx, color=colorx, label='X')
    #Se grafican los datos de y
    plt.bar(inty, histy, color=colory, label='Y')
    plt.title('Comparación de histogramas', size=25, color='mediumorchid')
    plt.xticks(color='slategrey')
    plt.yticks(color='slategrey')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def grafica(x, title='', color='lightcoral'):
    plt.figure(figsize=(15,5))
    plt.plot(x, color=color)
    plt.title(title, color='crimson', size=25)
    plt.xticks(color='slategrey')
    plt.yticks(color='slategrey')
    plt.grid()
    plt.show()
```

### Definimos una variable aleatoria X con distribución Normal, 
### $\mu$ = 0 y $\sigma$ = .1

```python
x = np.random.normal(0, .1, 1000)

grafica(x, 'Variable aleatoria X')
int_x, hist_x = histogram(x, title='Histograma de X', width=.1, color='deeppink')
```

 Definimos la transformación como
 $$ y = ce^{-\alpha x}$$
 donde $c$ y $\alpha$ son constantes

```python
def trans(x, c=1, alpha=1):
    #Para fines prácticos, los valores por default de alfa y de c, son 1
    return c*np.exp(-alpha*x)
```

Calculamos Y con $c = 1$ y $\alpha = 1$

```python
y = trans(x)

grafica(y, 'Variable aleatoria Y')
int_y, hist_y = histogram(y, nbins=np.sqrt(y.shape[0]), title='Histograma de Y', width=.1)
graph2(int_x, hist_x, int_y, hist_y)
```

Calculamos Y con $c = 1$ y $\alpha = 10$

```python
y = trans(x, alpha=10)

grafica(y, 'Variable aleatoria Y')
int_y, hist_y = histogram(y, nbins=np.sqrt(y.shape[0]), title='Histograma de Y', width=1)
graph2(int_x, hist_x, int_y, hist_y)
```

Calculamos Y con $c = 1$ y $\alpha = .01$

```python
y = trans(x, alpha=.01)

grafica(y, 'Variable aleatoria Y')
int_y, hist_y = histogram(y, nbins=np.sqrt(y.shape[0]), title='Histograma de Y', width=.05)
graph2(int_x, hist_x, int_y, hist_y)
```
