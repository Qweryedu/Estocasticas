```python
#Bibliotecas
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Se define A con una distribucin N
A = np.random.normal(0,1,size=1000)

#Se define el intervalo n
n = np.linspace(-10,10,500)
#Se define la frecuencia
f0 = .25

#Definimos la figura
plt.figure('Grafica', figsize=(15,5))

x = []

#Se define la funcin X
for a in A:
    temp = a*np.cos(2*np.pi*f0*n)
    x.append(temp)
    plt.plot(temp)
plt.plot(n,np.ones(n.shape)*3)
plt.plot(n,-3*np.ones(n.shape))
```
